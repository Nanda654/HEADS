```python
# @title Cell 1: ConstrainedBartSummarizer Approach
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import (
    BartForConditionalGeneration,
    BartTokenizer,
    LogitsProcessorList,
    MinLengthLogitsProcessor,
    BeamSearchScorer
)
import spacy
import numpy as np
from typing import List, Dict, Tuple, Set, Optional


class ConstraintExtractor:
    """Extract factual constraints from key sentences."""

    def __init__(self):
        try:
            self.nlp = spacy.load("en_core_web_lg")
        except OSError:
            print("Downloading spaCy model 'en_core_web_lg'...")
            import os
            os.system("python -m spacy download en_core_web_lg")
            self.nlp = spacy.load("en_core_web_lg")


    def extract_constraints(self, key_sentences: List[str]) -> List[Dict]:
        """
        Extract named entities, dates, numbers, and other factual elements
        from key sentences to use as constraints.
        """
        constraints = []

        for sentence in key_sentences:
            doc = self.nlp(sentence)

            # Extract named entities
            entities = [(ent.text, ent.label_) for ent in doc.ents]

            # Extract numerical values
            numbers = []
            for token in doc:
                if token.like_num and not any(token.text in e[0] for e in entities):
                    numbers.append(token.text)

            # Extract key noun phrases
            noun_chunks = [chunk.text for chunk in doc.noun_chunks
                          if not any(chunk.text in e[0] for e in entities)]

            # Create constraint dictionary
            constraint = {
                "sentence": sentence,
                "entities": entities,
                "numbers": numbers,
                "noun_chunks": noun_chunks
            }

            constraints.append(constraint)

        return constraints

    def format_constraints_for_bart(self, constraints: List[Dict]) -> str:
        """Format constraints as a string to prepend to BART input."""
        formatted = "Important facts to include:\n"

        for i, constraint in enumerate(constraints):
            formatted += f"[Fact {i+1}] "

            # Add entities
            if constraint["entities"]:
                entities_str = ", ".join([f"{e[0]} ({e[1]})" for e in constraint["entities"]])
                formatted += f"Entities: {entities_str}. "

            # Add numbers
            if constraint["numbers"]:
                numbers_str = ", ".join(constraint["numbers"])
                formatted += f"Numbers: {numbers_str}. "

            # Add key phrases
            if constraint["noun_chunks"]:
                phrases_str = ", ".join(constraint["noun_chunks"][:3])  # Limit to top 3
                formatted += f"Key phrases: {phrases_str}."

            formatted += "\n"

        return formatted

    def get_constraint_tokens(self, constraints: List[Dict], tokenizer) -> Set[int]:
        """Extract important tokens from constraints to guide beam search."""
        important_words = set()

        for constraint in constraints:
            # Add entity texts
            for entity, _ in constraint["entities"]:
                important_words.add(entity.lower())
                # Add individual words from multi-word entities
                important_words.update(entity.lower().split())

            # Add numbers
            important_words.update([num.lower() for num in constraint["numbers"]])

            # Add key noun phrases
            for chunk in constraint["noun_chunks"]:
                important_words.add(chunk.lower())
                # Add individual words from chunks
                important_words.update(chunk.lower().split())

        # Convert words to token IDs
        token_ids = set()
        for word in important_words:
            tokens = tokenizer.encode(word, add_special_tokens=False)
            token_ids.update(tokens)

        return token_ids


class ConstrainedLogitsProcessor(nn.Module):
    """Logits processor that boosts the probability of constraint tokens."""

    def __init__(self, constraint_token_ids: Set[int], boost_factor: float = 2.0):
        super().__init__()
        # Convert constraint_token_ids to a tensor and keep it on the correct device
        self.constraint_token_ids = torch.tensor(list(constraint_token_ids), dtype=torch.long) # Remove .to(device) here
        self.boost_factor = boost_factor

    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor) -> torch.FloatTensor:
        """Boost scores for constraint tokens."""
        # Ensure scores tensor is on the correct device
        # Move constraint_token_ids to the device of scores inside __call__
        constraint_token_ids_on_device = self.constraint_token_ids.to(scores.device)

        # Check if constraint token IDs are within the vocabulary size
        max_vocab_id = scores.size(1)
        valid_constraint_token_ids = constraint_token_ids_on_device[constraint_token_ids_on_device < max_vocab_id]

        # Apply boost to valid constraint tokens
        scores[:, valid_constraint_token_ids] *= self.boost_factor

        return scores


class ConstrainedBartSummarizer:
    """BART summarizer with constrained beam search."""

    def __init__(self, model_name: str = "facebook/bart-large-cnn"):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.tokenizer = BartTokenizer.from_pretrained(model_name)
        self.model = BartForConditionalGeneration.from_pretrained(model_name).to(self.device)
        self.constraint_extractor = ConstraintExtractor()

    def summarize_with_constraints(
        self,
        document: str,
        key_sentences: List[str],
        use_constraint_prefix: bool = True,
        use_constrained_beam_search: bool = True,
        num_beams: int = 4,
        min_length_pct: float = 0.75,
        max_length_pct: float = 0.9,
        boost_factor: float = 2.0,
        no_repeat_ngram_size: int = 3
    ) -> str:
        """Generate a summary using constraints from key sentences."""
        # Calculate min and max length based on document length
        doc_length = len(document.split())
        min_length = int(doc_length * min_length_pct)
        max_length = int(doc_length * max_length_pct)

        print(f"Document length: {doc_length} words")
        print(f"Summary length range: {min_length}-{max_length} words")

        # Extract constraints from key sentences
        constraints = self.constraint_extractor.extract_constraints(key_sentences)

        # Prepare input text
        input_text = document
        if use_constraint_prefix:
            constraint_prefix = self.constraint_extractor.format_constraints_for_bart(constraints)
            input_text = constraint_prefix + "\n\nDocument: " + document

        # Tokenize input
        inputs = self.tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        # Set up generation parameters
        generation_kwargs = {
            "num_beams": num_beams,
            "min_length": min_length,
            "max_length": max_length,
            "no_repeat_ngram_size": no_repeat_ngram_size,
            "early_stopping": True,
        }

        # Apply constrained beam search if requested
        if use_constrained_beam_search:
            print("Using constrained beam search in ConstrainedBartSummarizer...")
            constraint_token_ids = self.constraint_extractor.get_constraint_tokens(
                constraints, self.tokenizer
            )

            # Create logits processors
            logits_processor = LogitsProcessorList([
                MinLengthLogitsProcessor(min_length, self.tokenizer.eos_token_id),
                ConstrainedLogitsProcessor(constraint_token_ids, boost_factor)
            ])

            generation_kwargs["logits_processor"] = logits_processor
        else:
             print("Using standard beam search in ConstrainedBartSummarizer...")

        # Generate summary
        with torch.no_grad():
            output_ids = self.model.generate(**inputs, **generation_kwargs)

        # Decode and return summary
        summary = self.tokenizer.decode(output_ids[0], skip_special_tokens=True)
        return summary


# # Function to run the ConstrainedBartSummarizer
# def run_constrained_bart_summarizer(document, extractive_key_sentences):
#     """
#     Run the ConstrainedBartSummarizer.

#     Args:
#         document: The full document text
#         extractive_key_sentences: Key sentences extracted by Longformer

#     Returns:
#         Generated summary
#     """
#     print("\n--- Running ConstrainedBartSummarizer ---")
#     summarizer = ConstrainedBartSummarizer()
#     summary = summarizer.summarize_with_constraints(
#         document=document,
#         key_sentences=extractive_key_sentences,
#         use_constraint_prefix=True,
#         use_constrained_beam_search=True,
#         min_length_pct=0.75,
#         max_length_pct=0.9
#     )
#     print("--- Finished ConstrainedBartSummarizer ---\n")

#     return summary
```

```python
# @title Cell 2: ConstraintGuidedBART Approach
import torch
import torch.nn as nn
from transformers import (
    BartForConditionalGeneration,
    BartTokenizer,
    LogitsProcessorList,
    MinLengthLogitsProcessor
)
import spacy
from typing import List, Dict, Set


class ConstraintExtractor:
    """Extract factual constraints from key sentences."""

    def __init__(self):
        try:
            self.nlp = spacy.load("en_core_web_lg")
        except OSError:
            print("Downloading spaCy model 'en_core_web_lg'...")
            import os
            os.system("python -m spacy download en_core_web_lg")
            self.nlp = spacy.load("en_core_web_lg")

    def extract_constraints(self, key_sentences: List[str]) -> List[Dict]:
        """
        Extract named entities, dates, numbers, and other factual elements
        from key sentences to use as constraints.
        """
        constraints = []

        for sentence in key_sentences:
            doc = self.nlp(sentence)

            # Extract named entities
            entities = [(ent.text, ent.label_) for ent in doc.ents]

            # Extract numerical values
            numbers = []
            for token in doc:
                if token.like_num and not any(token.text in e[0] for e in entities):
                    numbers.append(token.text)

            # Extract key noun phrases
            noun_chunks = [chunk.text for chunk in doc.noun_chunks
                          if not any(chunk.text in e[0] for e in entities)]

            # Create constraint dictionary
            constraint = {
                "sentence": sentence,
                "entities": entities,
                "numbers": numbers,
                "noun_chunks": noun_chunks
            }

            constraints.append(constraint)

        return constraints

    def get_constraint_tokens(self, constraints: List[Dict], tokenizer) -> Set[int]:
        """Extract important tokens from constraints to guide beam search."""
        important_words = set()

        for constraint in constraints:
            # Add entity texts
            for entity, _ in constraint["entities"]:
                important_words.add(entity.lower())
                # Add individual words from multi-word entities
                important_words.update(entity.lower().split())

            # Add numbers
            important_words.update([num.lower() for num in constraint["numbers"]])

            # Add key noun phrases
            for chunk in constraint["noun_chunks"]:
                important_words.add(chunk.lower())
                # Add individual words from chunks
                important_words.update(chunk.lower().split())

        # Convert words to token IDs
        token_ids = set()
        for word in important_words:
            tokens = tokenizer.encode(word, add_special_tokens=False)
            token_ids.update(tokens)

        return token_ids


class ConstrainedLogitsProcessor(nn.Module):
    """Logits processor that boosts the probability of constraint tokens."""

    def __init__(self, constraint_token_ids: Set[int], boost_factor: float = 2.0):
        super().__init__()
        # Convert constraint_token_ids to a tensor and keep it on the correct device
        self.constraint_token_ids = torch.tensor(list(constraint_token_ids), dtype=torch.long) # Remove .to(device) here
        self.boost_factor = boost_factor

    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor) -> torch.FloatTensor:
        """Boost scores for constraint tokens."""
        # Ensure scores tensor is on the correct device
        # Move constraint_token_ids to the device of scores inside __call__
        constraint_token_ids_on_device = self.constraint_token_ids.to(scores.device)

        # Check if constraint token IDs are within the vocabulary size
        max_vocab_id = scores.size(1)
        valid_constraint_token_ids = constraint_token_ids_on_device[constraint_token_ids_on_device < max_vocab_id]

        # Apply boost to valid constraint tokens
        scores[:, valid_constraint_token_ids] *= self.boost_factor

        return scores


class ConstraintGuidedBART(nn.Module):
    """BART model with direct constraint integration."""

    def __init__(self, model_name="facebook/bart-large-cnn"):
        super().__init__()
        self.model = BartForConditionalGeneration.from_pretrained(model_name)
        self.tokenizer = BartTokenizer.from_pretrained(model_name)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model.to(self.device)

    def forward(self, input_ids, attention_mask, decoder_input_ids=None, labels=None):
        """Forward pass with standard BART."""
        return self.model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            decoder_input_ids=decoder_input_ids,
            labels=labels
        )

    def generate_with_constraints(
        self,
        input_ids,
        attention_mask,
        constraint_token_ids,
        num_beams=4,
        min_length=400,
        max_length=500,
        boost_factor=2.0
    ):
        """Generate text with constraint-guided beam search."""
        # Create logits processor with constraints
        logits_processor = LogitsProcessorList([
            MinLengthLogitsProcessor(min_length, self.tokenizer.eos_token_id),
            ConstrainedLogitsProcessor(constraint_token_ids, boost_factor)
        ])

        # Generate with constraints
        outputs = self.model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            num_beams=num_beams,
            min_length=min_length,
            max_length=max_length,
            logits_processor=logits_processor,
            early_stopping=True,
            no_repeat_ngram_size=3
        )

        return outputs

    def summarize(self, document, key_sentences, min_length_pct=0.75, max_length_pct=0.9):
        """Generate summary with constraints from key sentences."""
        # Calculate min and max length based on document length
        doc_length = len(document.split())
        min_length = int(doc_length * min_length_pct)
        max_length = int(doc_length * max_length_pct)

        print(f"Document length: {doc_length} words")
        print(f"Summary length range: {min_length}-{max_length} words")

        # Extract constraints
        constraint_extractor = ConstraintExtractor()
        constraints = constraint_extractor.extract_constraints(key_sentences)
        constraint_token_ids = constraint_extractor.get_constraint_tokens(constraints, self.tokenizer)

        # Tokenize input
        inputs = self.tokenizer(document, return_tensors="pt", max_length=1024, truncation=True)
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        # Generate summary
        print("Using constraint-guided beam search in ConstraintGuidedBART...")
        output_ids = self.generate_with_constraints(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            constraint_token_ids=constraint_token_ids,
            min_length=min_length,
            max_length=max_length
        )

        # Decode summary
        summary = self.tokenizer.decode(output_ids[0], skip_special_tokens=True)
        return summary


# Function to run the ConstraintGuidedBART
def run_constraint_guided_bart(document, extractive_key_sentences):
    """
    Run the ConstraintGuidedBART.

    Args:
        document: The full document text
        extractive_key_sentences: Key sentences extracted by Longformer

    Returns:
        Generated summary
    """
    print("\n--- Running ConstraintGuidedBART ---")
    summarizer = ConstraintGuidedBART()
    summary = summarizer.summarize(
        document=document,
        key_sentences=extractive_key_sentences,
        min_length_pct=0.75,
        max_length_pct=0.9
    )
    print("--- Finished ConstraintGuidedBART ---\n")

    return summary
```

```python
# @title Bart with Dynamic Length Constraints

# Install necessary libraries (only runs if not already installed)
#%pip install transformers torch

import torch
from transformers import BartForConditionalGeneration, BartTokenizer

# --- Load BART Model and Tokenizer ---
print("Loading BART model and tokenizer for abstractive summarization...")
bart_model_name = 'facebook/bart-large-cnn' # This is a good choice for summarization
bart_tokenizer = BartTokenizer.from_pretrained(bart_model_name)
bart_model = BartForConditionalGeneration.from_pretrained(bart_model_name)

# --- Set Device (GPU if available, else CPU) ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

bart_model.eval() # Set to evaluation mode
bart_model.to(device)
print(f"BART using device: {device}")
print("BART model loaded.")

# --- Abstractive Summarization Function (using BART) ---
def bart_abstractive_summary(text_to_summarize, min_length_pct=0.75, max_length_pct=0.9, num_beams=5, early_stopping=True):
    """
    Generates an abstractive summary using the pre-loaded BART model.
    Summary length is dynamically set based on input document length.

    Args:
        text_to_summarize (str or list of str): The input text (or list of sentences) to summarize.
                                                If a list, it will be joined into a single string.
        min_length_pct (float): Minimum length of summary as a percentage of input document length.
        max_length_pct (float): Maximum length of summary as a percentage of input document length.
        num_beams (int): Number of beams for beam search. Higher values lead to better quality but slower generation.
        early_stopping (bool): Whether to stop beam search when all beams have finished their generation.

    Returns:
        str: The generated abstractive summary.
    """
    print("\n--- Starting BART Abstractive Summarization ---")

    if isinstance(text_to_summarize, list):
        text_to_summarize = " ".join(text_to_summarize)

    if not text_to_summarize.strip():
        print("  Input text for abstractive summary is empty. Cannot summarize.")
        return ""

    # Calculate dynamic length constraints based on input document length
    doc_length = len(text_to_summarize.split())
    min_length = max(50, int(doc_length * min_length_pct))  # Ensure minimum is at least 50 tokens
    max_length = min(1024, int(doc_length * max_length_pct))  # Ensure maximum doesn't exceed model limits

    print(f"Document length: {doc_length} words")
    print(f"Summary length range: {min_length}-{max_length} words")

    inputs = bart_tokenizer(
        [text_to_summarize],
        max_length=1024,  # BART's typical max input length
        return_tensors="pt",
        truncation=True,
        padding="max_length"
    ).to(device)

    summary_ids = bart_model.generate(
        inputs["input_ids"],
        num_beams=num_beams,
        max_length=max_length,
        min_length=min_length,
        early_stopping=early_stopping
    )

    summary_text = bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Print actual summary length for verification
    summary_length = len(summary_text.split())
    print(f"Generated summary length: {summary_length} words")
    print("--- BART Abstractive Summarization Complete ---")

    return summary_text

# # --- Example Usage for BART only ---
# print("\n" + "="*80)
# print("Abstractive Summary (using BART directly on the full text):")
# bart_only_summary = bart_abstractive_summary(
#     input_text_for_bart_1,
#     min_length_pct=0.75,  # Summary will be at least 75% of document length
#     max_length_pct=0.9,   # Summary will be at most 90% of document length
#     num_beams=5           # Beam search parameter for quality
# )
# print(bart_only_summary)

# print("\n" + "="*80)
# print("\nBART only summarization complete.")
```

```python
# @title evaluation example

import spacy
from sentence_transformers import SentenceTransformer, util
from rouge_score import rouge_scorer
import bert_score

# Load models
nlp = spacy.load("en_core_web_sm")
model = SentenceTransformer("all-MiniLM-L6-v2")

# Sentence splitter
def split_sentences(text):
    return [sent.text.strip() for sent in nlp(text).sents]

# Coverage score
def get_sentence_coverage(para1, para2, threshold=0.5):
    sents1 = split_sentences(para1)
    sents2 = split_sentences(para2)

    emb1 = model.encode(sents1, convert_to_tensor=True)
    emb2 = model.encode(sents2, convert_to_tensor=True)

    matched = 0
    for i in range(len(sents2)):
        sims = util.cos_sim(emb2[i], emb1)[0]
        if sims.max().item() >= threshold:
            matched += 1

    coverage = matched / len(sents2) if sents2 else 0.0
    return round(coverage, 4)

# ROUGE score
def get_rouge_score(reference, candidate):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, candidate)
    return {
        "rouge1": round(scores["rouge1"].fmeasure, 4),
        "rouge2": round(scores["rouge2"].fmeasure, 4),
        "rougeL": round(scores["rougeL"].fmeasure, 4)
    }

# BERTScore
def get_bert_score(reference, candidate, lang="en"):
    P, R, F1 = bert_score.score([candidate], [reference], lang=lang, verbose=False)
    return {
        "bertscore_precision": round(P[0].item(), 4),
        "bertscore_recall": round(R[0].item(), 4),
        "bertscore_f1": round(F1[0].item(), 4)
    }

# Combined comparison
def compare_paragraphs(para1, para2):
    scores = {
        "coverage_score": get_sentence_coverage(para1, para2),
        **get_rouge_score(para1, para2),
        **get_bert_score(para1, para2)
    }
    return scores
```

```python
# @title Cell: Key Sentence Extraction from Extractive Summary (5% limit)
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import sent_tokenize
import re
import math
from typing import List, Dict, Tuple

# Download necessary NLTK resources
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')


class KeySentenceExtractor:
    """Extract key sentences from an extractive summary based on importance scoring."""

    def __init__(self, max_percent=25, diversity_penalty=0.5):
        """
        Initialize the key sentence extractor.

        Args:
            max_percent: Maximum percentage of sentences to extract (default: 5%)
            diversity_penalty: Penalty factor for redundant information
        """
        self.max_percent = max_percent
        self.diversity_penalty = diversity_penalty
        self.vectorizer = TfidfVectorizer(
            lowercase=True,
            stop_words='english',
            ngram_range=(1, 2),
            max_features=5000
        )

    def clean_text(self, text: str) -> str:
        """Clean text by removing extra whitespace and special characters."""
        text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with single space
        text = re.sub(r'[^\w\s.,;:!?-]', '', text)  # Remove special characters except punctuation
        return text.strip()

    def extract_key_sentences(self, extractive_summary: str) -> List[str]:
        """
        Extract key sentences from the extractive summary.

        Args:
            extractive_summary: The extractive summary text

        Returns:
            List of key sentences
        """
        # Clean the text
        clean_summary = self.clean_text(extractive_summary)

        # Split into sentences
        sentences = sent_tokenize(clean_summary)

        # Calculate max sentences (5% of total sentences, minimum 1)
        max_sentences = max(1, math.ceil(len(sentences) * self.max_percent / 100))
        print(f"Total sentences in extractive summary: {len(sentences)}")
        print(f"Extracting {max_sentences} key sentences ({self.max_percent}% of total)")

        if len(sentences) <= max_sentences:
            return sentences

        # Calculate TF-IDF vectors for sentences
        tfidf_matrix = self.vectorizer.fit_transform(sentences)

        # Convert to dense array for easier manipulation
        dense_matrix = tfidf_matrix.toarray()

        # Calculate sentence importance scores
        sentence_scores = self._calculate_sentence_scores(dense_matrix)

        # Select top sentences using MMR (Maximal Marginal Relevance)
        selected_indices = self._select_sentences_mmr(dense_matrix, sentence_scores, max_sentences)

        # Sort indices to maintain original order
        selected_indices.sort()

        # Return selected sentences
        return [sentences[i] for i in selected_indices]

    def _calculate_sentence_scores(self, tfidf_matrix: np.ndarray) -> np.ndarray:
        """
        Calculate importance score for each sentence based on TF-IDF values.

        Args:
            tfidf_matrix: TF-IDF matrix of sentences

        Returns:
            Array of sentence scores
        """
        # Sum TF-IDF values for each sentence
        return np.sum(tfidf_matrix, axis=1)

    def _select_sentences_mmr(self, tfidf_matrix: np.ndarray, sentence_scores: np.ndarray, max_sentences: int) -> List[int]:
        """
        Select sentences using Maximal Marginal Relevance to reduce redundancy.

        Args:
            tfidf_matrix: TF-IDF matrix of sentences
            sentence_scores: Importance score for each sentence
            max_sentences: Maximum number of sentences to select

        Returns:
            Indices of selected sentences
        """
        num_sentences = tfidf_matrix.shape[0]
        selected_indices = []
        unselected_indices = list(range(num_sentences))

        # Calculate similarity matrix
        similarity_matrix = np.dot(tfidf_matrix, tfidf_matrix.T)

        # Normalize similarity scores
        norms = np.sqrt(np.diag(similarity_matrix))
        for i in range(num_sentences):
            for j in range(num_sentences):
                if norms[i] > 0 and norms[j] > 0:
                    similarity_matrix[i, j] /= (norms[i] * norms[j])

        # Select first sentence with highest score
        first_idx = np.argmax(sentence_scores)
        selected_indices.append(first_idx)
        unselected_indices.remove(first_idx)

        # Select remaining sentences using MMR
        while len(selected_indices) < max_sentences and unselected_indices:
            mmr_scores = np.zeros(len(unselected_indices))

            for i, idx in enumerate(unselected_indices):
                # Calculate maximum similarity to already selected sentences
                if selected_indices:
                    max_sim = max(similarity_matrix[idx, j] for j in selected_indices)
                else:
                    max_sim = 0

                # MMR score = relevance - diversity_penalty * redundancy
                mmr_scores[i] = sentence_scores[idx] - self.diversity_penalty * max_sim

            # Select sentence with highest MMR score
            next_idx = unselected_indices[np.argmax(mmr_scores)]
            selected_indices.append(next_idx)
            unselected_indices.remove(next_idx)

        return selected_indices

def extract_key_sentences_from_extractive(extractive_summary: str, max_percent: int =25) -> List[str]:
    """
    Extract key sentences from an extractive summary.

    Args:
        extractive_summary: The extractive summary text
        max_percent: Maximum percentage of sentences to extract (default: 5%)

    Returns:
        List of key sentences
    """
    extractor = KeySentenceExtractor(max_percent=max_percent)
    return extractor.extract_key_sentences(extractive_summary)

# Function to integrate with the ConstraintGuidedBART
def run_hybrid_summarization_pipeline(document: str, extractive_summary: str, max_percent: int =25) -> str:
    """
    Run the complete hybrid summarization pipeline.

    Args:
        document: The full document text
        extractive_summary: The extractive summary text
        max_percent: Maximum percentage of sentences to extract as key sentences

    Returns:
        Generated abstractive summary
    """
    print("\n--- Extracting Key Sentences ---")
    key_sentences = extract_key_sentences_from_extractive(extractive_summary, max_percent)

    print("\nExtracted Key Sentences:")
    for i, sentence in enumerate(key_sentences, 1):
        print(f"{i}. {sentence}")

    # Run the constraint-guided BART with the extracted key sentences
    abstractive_summary = run_constraint_guided_bart(document, key_sentences)

    return abstractive_summary

# Example usage
if __name__ == "__main__":
    # Sample extractive summary (replace with your actual extractive summary)
    extractive_summary = input_text_for_bart_1

    # Sample document (in a real scenario, this would be the full document)
    document = "This is a placeholder for the full document text. " + extractive_summary * 3

    # Extract key sentences and run the full pipeline
    key_sentences = extract_key_sentences_from_extractive(input_text_for_bart_1)

    print("\nExtracted Key Sentences:")
    for i, sentence in enumerate(key_sentences, 1):
        print(f"{i}. {sentence}")

    # Uncomment to run the full pipeline
    # abstractive_summary = run_hybrid_summarization_pipeline(document, extractive_summary)
    # print(f"\nAbstractive Summary:\n{abstractive_summary}")
```

```python
# @title Cell 1: Load and Evaluate Models on Dataset_-0

import json
import torch
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from google.colab import drive
import os

# Mount Google Drive to access the dataset
drive.mount('/content/drive')

# Function to load data from JSON file
def load_data(file_path, num_samples=None):
    """
    Load data from JSON file.

    Args:
        file_path: Path to the JSON file
        num_samples: Number of samples to load (None for all)

    Returns:
        List of dictionaries containing the data
    """
    print(f"Loading data from {file_path}...")
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    if isinstance(data, dict):
        # Convert to list if it's a dictionary
        data = [data]

    if num_samples is not None:
        data = data[:num_samples]

    print(f"Loaded {len(data)} samples.")
    return data

# Function to evaluate both models on the dataset
def evaluate_models(data, num_samples=None):
    """
    Evaluate both models on the dataset.

    Args:
        data: List of dictionaries containing the data
        num_samples: Number of samples to evaluate (None for all)

    Returns:
        DataFrame with evaluation results
    """
    if num_samples is not None:
        data = data[:num_samples]

    results = []

    for i, sample in enumerate(tqdm(data, desc="Evaluating samples")):
        print(f"\n{'='*80}\nProcessing sample {i+1}/{len(data)}\n{'='*80}")

        original_text = sample.get("original_text", "")
        extractive_summary = sample.get("extractive_summary", "")
        ground_truth = sample.get("abstractive_summary", "")

        if not original_text or not extractive_summary or not ground_truth:
            print(f"Skipping sample {i+1} due to missing data")
            continue

        # Convert extractive_summary to list if it's a string
        extractive_key_sentences = extractive_summary
        if isinstance(extractive_summary, str):
            extractive_key_sentences = split_sentences(extractive_summary)

        # Generate summaries using both models
        print("\n--- Generating summaries ---")

        # Standard BART
        print("\nGenerating summary with standard BART...")
        bart_summary = bart_abstractive_summary(
            original_text,
            min_length_pct=0.75,
            max_length_pct=0.9,
            num_beams=4
        )

        # ConstraintGuidedBART
        print("\nGenerating summary with ConstraintGuidedBART...")
        constraint_guided_summary = run_constraint_guided_bart(
            original_text,
            extractive_key_sentences
        )

        # Evaluate both summaries against ground truth
        print("\n--- Evaluating summaries ---")

        print("\nEvaluating standard BART summary...")
        bart_scores = compare_paragraphs(ground_truth, bart_summary)

        print("\nEvaluating ConstraintGuidedBART summary...")
        constraint_guided_scores = compare_paragraphs(ground_truth, constraint_guided_summary)

        # Store results
        result = {
            "sample_id": i+1,
            "bart_coverage": bart_scores["coverage_score"],
            "bart_rouge1": bart_scores["rouge1"],
            "bart_rouge2": bart_scores["rouge2"],
            "bart_rougeL": bart_scores["rougeL"],
            "bart_bertscore_f1": bart_scores["bertscore_f1"],
            "constraint_coverage": constraint_guided_scores["coverage_score"],
            "constraint_rouge1": constraint_guided_scores["rouge1"],
            "constraint_rouge2": constraint_guided_scores["rouge2"],
            "constraint_rougeL": constraint_guided_scores["rougeL"],
            "constraint_bertscore_f1": constraint_guided_scores["bertscore_f1"],
            "original_text_length": len(original_text.split()),
            "ground_truth_length": len(ground_truth.split()),
            "bart_summary_length": len(bart_summary.split()),
            "constraint_summary_length": len(constraint_guided_summary.split()),
            "original_text": original_text[:500] + "...",  # Truncated for display
            "ground_truth": ground_truth,
            "bart_summary": bart_summary,
            "constraint_summary": constraint_guided_summary
        }

        results.append(result)

    # Convert to DataFrame
    df = pd.DataFrame(results)
    return df

# Function to visualize results
def visualize_results(df):
    """
    Visualize evaluation results.

    Args:
        df: DataFrame with evaluation results
    """
    # Calculate average scores
    avg_scores = {
        "BART": {
            "Coverage": df["bart_coverage"].mean(),
            "ROUGE-1": df["bart_rouge1"].mean(),
            "ROUGE-2": df["bart_rouge2"].mean(),
            "ROUGE-L": df["bart_rougeL"].mean(),
            "BERTScore F1": df["bart_bertscore_f1"].mean()
        },
        "ConstraintGuidedBART": {
            "Coverage": df["constraint_coverage"].mean(),
            "ROUGE-1": df["constraint_rouge1"].mean(),
            "ROUGE-2": df["constraint_rouge2"].mean(),
            "ROUGE-L": df["constraint_rougeL"].mean(),
            "BERTScore F1": df["constraint_bertscore_f1"].mean()
        }
    }

    # Print average scores
    print("\n--- Average Scores ---")
    for model, scores in avg_scores.items():
        print(f"\n{model}:")
        for metric, score in scores.items():
            print(f"  {metric}: {score:.4f}")

    # Create bar chart
    metrics = ["Coverage", "ROUGE-1", "ROUGE-2", "ROUGE-L", "BERTScore F1"]
    bart_scores = [avg_scores["BART"][m] for m in metrics]
    constraint_scores = [avg_scores["ConstraintGuidedBART"][m] for m in metrics]

    x = range(len(metrics))
    width = 0.35

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar([i - width/2 for i in x], bart_scores, width, label='BART')
    ax.bar([i + width/2 for i in x], constraint_scores, width, label='ConstraintGuidedBART')

    ax.set_ylabel('Score')
    ax.set_title('Model Performance Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()

    plt.tight_layout()
    plt.show()

    # Create length comparison
    fig, ax = plt.subplots(figsize=(10, 6))

    lengths = {
        "Ground Truth": df["ground_truth_length"].mean(),
        "BART Summary": df["bart_summary_length"].mean(),
        "ConstraintGuided Summary": df["constraint_summary_length"].mean()
    }

    ax.bar(lengths.keys(), lengths.values())
    ax.set_ylabel('Average Word Count')
    ax.set_title('Summary Length Comparison')

    plt.tight_layout()
    plt.show()

# Main execution function
def run_evaluation(file_path, num_samples=1):
    """
    Run the evaluation pipeline.

    Args:
        file_path: Path to the JSON file
        num_samples: Number of samples to evaluate
    """
    # Load data
    data = load_data(file_path, num_samples)

    # Evaluate models
    results_df = evaluate_models(data, num_samples)

    # Visualize results
    visualize_results(results_df)

    # Save results to CSV
    csv_path = os.path.join(os.path.dirname(file_path), 'evaluation_results.csv')
    results_df.to_csv(csv_path, index=False)
    print(f"\nResults saved to {csv_path}")

    return results_df

# Example usage
# Specify the path to your JSON file and the number of samples to evaluate
file_path = '/content/drive/MyDrive/your_data_file.json'  # Update this path
num_samples = 1  # Change this to evaluate more samples

# Run the evaluation
results = run_evaluation(file_path, num_samples)
```

```python
import json
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from google.colab import drive
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from bert_score import score as bert_score
import nltk
import os
import psutil
from typing import Dict, List, Any, Tuple
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# ===================== DEVICE SETUP =====================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================== MEMORY CONFIG =====================
def configure_memory_usage():
    try:
        mem_info = psutil.virtual_memory()
        total_ram_gb = mem_info.total / (1024**3)
        available_ram_gb = mem_info.available / (1024**3)

        print(f"Total RAM: {total_ram_gb:.2f} GB")
        print(f"Available RAM: {available_ram_gb:.2f} GB")

        if 'google.colab' in str(get_ipython()):
            print("Note: For high-RAM runtime in Colab: Runtime > Change runtime type > Hardware accelerator")

        if available_ram_gb > 4:
            torch.cuda.empty_cache()
            if torch.cuda.is_available():
                torch.cuda.set_per_process_memory_fraction(0.8)
                print(f"Set PyTorch to use up to 80% of GPU memory")
    except Exception as e:
        print(f"Error configuring memory: {e}")

configure_memory_usage()

# ===================== NLTK SETUP =====================
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

# ===================== MOUNT DRIVE =====================
print("Mounting Google Drive...")
drive.mount('/content/drive')

# ===================== FACTCC EVALUATOR =====================
class FactCCEvaluator:
    def __init__(self, model_name="manueldeprada/factcc", device=None):
        self.device = device if device is not None else torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print("Loading FactCC model...")
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForSequenceClassification.from_pretrained(model_name).to(self.device)
        self.model.eval()

    def evaluate(self, source_text: str, generated_summary: str) -> float:
        inputs = self.tokenizer(
            source_text,
            generated_summary,
            return_tensors="pt",
            truncation=True,
            padding="longest"
        ).to(self.device)

        with torch.no_grad():
            outputs = self.model(**inputs)
            probs = torch.softmax(outputs.logits, dim=1)
            consistency_prob = probs[0][0].item()  # label 0 = CONSISTENT
        return consistency_prob

# ===================== SUMMARIZATION EVALUATOR =====================
class SummarizationEvaluator:
    def __init__(self, device=None):
        self.rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
        self.smoothing = SmoothingFunction().method1
        self.device = device if device is not None else torch.device("cuda" if torch.cuda.is_available() else "cpu")

    def evaluate_summary(self, generated_summary: str, reference_summary: str) -> Dict[str, float]:
        rouge_scores = self.rouge_scorer.score(reference_summary, generated_summary)
        reference_tokens = [reference_summary.split()]
        candidate_tokens = generated_summary.split()
        bleu_score_val = sentence_bleu(reference_tokens, candidate_tokens, smoothing_function=self.smoothing)

        if generated_summary.strip() and reference_summary.strip():
            try:
                P, R, F1 = bert_score([generated_summary], [reference_summary], lang='en', verbose=False, device='cpu')
                bert_p = P.item()
                bert_r = R.item()
                bert_f1 = F1.item()
            except Exception as e:
                print(f"Error calculating BERTScore: {e}")
                bert_p = bert_r = bert_f1 = 0.0
        else:
            bert_p = bert_r = bert_f1 = 0.0

        metrics = {
            'rouge1_precision': rouge_scores['rouge1'].precision,
            'rouge1_recall': rouge_scores['rouge1'].recall,
            'rouge1_f1': rouge_scores['rouge1'].fmeasure,
            'rouge2_precision': rouge_scores['rouge2'].precision,
            'rouge2_recall': rouge_scores['rouge2'].recall,
            'rouge2_f1': rouge_scores['rouge2'].fmeasure,
            'rougeL_precision': rouge_scores['rougeL'].precision,
            'rougeL_recall': rouge_scores['rougeL'].recall,
            'rougeL_f1': rouge_scores['rougeL'].fmeasure,
            'bleu': bleu_score_val,
            'bert_precision': bert_p,
            'bert_recall': bert_r,
            'bert_f1': bert_f1
        }
        return metrics

# ===================== DATA LOADING =====================
def load_data(file_path: str, num_samples: int = None) -> List[Dict[str, str]]:
    data = []
    print(f"Loading data from {file_path}...")
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line:
                try:
                    data.append(json.loads(line))
                except json.JSONDecodeError as e:
                    print(f"Skipping invalid JSON: {line[:100]}... Error: {e}")
            if num_samples is not None and len(data) >= num_samples:
                break
    print(f"Loaded {len(data)} samples.")
    return data

# ===================== MAIN EVALUATION =====================
def run_evaluation(data_path: str, num_samples: int = 1) -> Tuple[pd.DataFrame, Dict[str, Dict[str, float]]]:
    data = load_data(data_path, num_samples)
    evaluator = SummarizationEvaluator(device='cpu')
    factcc_evaluator = FactCCEvaluator(device='cpu')
    results = []

    for i, sample in enumerate(tqdm(data, desc="Evaluating samples")):
        original_text = sample.get("original_text", "")
        extractive_summary = sample.get("extractive_summary", "")
        reference_summary = sample.get("abstractive_summary", "")

        if not original_text or not extractive_summary or not reference_summary:
            continue

        try:
            key_sentences = extract_key_sentences_from_extractive(extractive_summary)
            constrained_summary = run_constraint_guided_bart(extractive_summary, key_sentences)
            standard_summary = bart_abstractive_summary(extractive_summary, min_length_pct=0.75, max_length_pct=0.9, num_beams=5)

            constrained_metrics = evaluator.evaluate_summary(constrained_summary, reference_summary)
            standard_metrics = evaluator.evaluate_summary(standard_summary, reference_summary)

            constrained_metrics['factcc_consistency'] = factcc_evaluator.evaluate(constrained_summary, extractive_summary)
            standard_metrics['factcc_consistency'] = factcc_evaluator.evaluate(standard_summary, extractive_summary)

            results.append({'sample_id': i, 'model': 'ConstraintGuidedBART', 'generated_summary': constrained_summary, 'reference_summary': reference_summary, **constrained_metrics})
            results.append({'sample_id': i, 'model': 'StandardBART', 'generated_summary': standard_summary, 'reference_summary': reference_summary, **standard_metrics})

        except Exception as e:
            print(f"Error processing sample {i+1}: {e}")
            continue
        finally:
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

    results_df = pd.DataFrame(results)
    avg_metrics = {}
    for model in ['ConstraintGuidedBART', 'StandardBART']:
        model_df = results_df[results_df['model'] == model]
        if not model_df.empty:
            avg_metrics[model] = {col: model_df[col].mean() for col in model_df.columns if col not in ['sample_id', 'model', 'generated_summary', 'reference_summary']}
        else:
            avg_metrics[model] = {}

    print("\nAverage Metrics:")
    for model, metrics in avg_metrics.items():
        if metrics:
            print(f"\n{model}:")
            print(f"  ROUGE-1 F1: {metrics.get('rouge1_f1', 'N/A')}")
            print(f"  ROUGE-2 F1: {metrics.get('rouge2_f1', 'N/A')}")
            print(f"  ROUGE-L F1: {metrics.get('rougeL_f1', 'N/A')}")
            print(f"  BLEU: {metrics.get('bleu', 'N/A')}")
            print(f"  BERTScore F1: {metrics.get('bert_f1', 'N/A')}")
            print(f"  FactCC Consistency: {metrics.get('factcc_consistency', 'N/A')}")

    return results_df, avg_metrics

# ===================== VISUALIZATION =====================
def visualize_results(avg_metrics: Dict[str, Dict[str, float]]):
    if not avg_metrics or all(not metrics for metrics in avg_metrics.values()):
        print("No metrics to visualize.")
        return

    metrics_to_plot = ['rouge1_f1', 'rouge2_f1', 'rougeL_f1', 'bleu', 'bert_f1', 'factcc_consistency']
    metric_names = ['ROUGE-1 F1', 'ROUGE-2 F1', 'ROUGE-L F1', 'BLEU', 'BERTScore F1', 'FactCC Consistency']

    models = [m for m in avg_metrics.keys() if avg_metrics[m]]
    x = np.arange(len(metric_names))
    width = 0.35
    fig, ax = plt.subplots(figsize=(12, 6))

    for i, model in enumerate(models):
        values = [avg_metrics[model].get(metric, 0) for metric in metrics_to_plot]
        ax.bar(x + i*width - width/2, values, width, label=model)
        for j, v in enumerate(values):
            ax.text(j + i*width - width/2, v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontsize=8)

    ax.set_ylabel('Score')
    ax.set_title('Summarization Model Performance')
    ax.set_xticks(x)
    ax.set_xticklabels(metric_names)
    ax.legend()
    plt.tight_layout()
    plt.show()

# ===================== MAIN FUNCTION =====================
def evaluate_models(data_path: str, num_samples: int = 1):
    results_df, avg_metrics = run_evaluation(data_path, num_samples)
    if not results_df.empty:
        visualize_results(avg_metrics)
        results_path = '/content/drive/MyDrive/summarization_results.csv'
        results_df.to_csv(results_path, index=False)
        print(f"\nResults saved to {results_path}")
    return results_df, avg_metrics

# ===================== EXAMPLE RUN =====================
if __name__ == "__main__":
    evaluate_models(data_path="/content/drive/MyDrive/MS_Fianl_Project/data1/validation.json", num_samples=1)
```

```python
import torch
import torch.nn as nn
from transformers import (
    BartForConditionalGeneration,
    BartTokenizer,
    LogitsProcessorList,
    MinLengthLogitsProcessor
)
import spacy
from typing import List, Dict, Set


class ConstraintExtractor:
    """Extract factual constraints from key sentences."""

    def __init__(self):
        try:
            self.nlp = spacy.load("en_core_web_lg")
        except OSError:
            print("Downloading spaCy model 'en_core_web_lg'...")
            import os
            os.system("python -m spacy download en_core_web_lg")
            self.nlp = spacy.load("en_core_web_lg")

    def extract_constraints(self, key_sentences: List[str]) -> List[Dict]:
        """
        Extract named entities, dates, numbers, and other factual elements
        from key sentences to use as constraints.
        """
        constraints = []

        for sentence in key_sentences:
            doc = self.nlp(sentence)

            # Extract named entities
            entities = [(ent.text, ent.label_) for ent in doc.ents]

            # Extract numerical values
            numbers = []
            for token in doc:
                if token.like_num and not any(token.text in e[0] for e in entities):
                    numbers.append(token.text)

            # Extract key noun phrases
            noun_chunks = [chunk.text for chunk in doc.noun_chunks
                          if not any(chunk.text in e[0] for e in entities)]

            # Create constraint dictionary
            constraint = {
                "sentence": sentence,
                "entities": entities,
                "numbers": numbers,
                "noun_chunks": noun_chunks
            }

            constraints.append(constraint)

        return constraints

    def get_constraint_tokens(self, constraints: List[Dict], tokenizer) -> Set[int]:
        """Extract important tokens from constraints to guide beam search."""
        important_words = set()

        for constraint in constraints:
            # Add entity texts
            for entity, _ in constraint["entities"]:
                important_words.add(entity.lower())
                # Add individual words from multi-word entities
                important_words.update(entity.lower().split())

            # Add numbers
            important_words.update([num.lower() for num in constraint["numbers"]])

            # Add key noun phrases
            for chunk in constraint["noun_chunks"]:
                important_words.add(chunk.lower())
                # Add individual words from chunks
                important_words.update(chunk.lower().split())

        # Convert words to token IDs
        token_ids = set()
        for word in important_words:
            tokens = tokenizer.encode(word, add_special_tokens=False)
            token_ids.update(tokens)

        return token_ids


class ConstrainedLogitsProcessor(nn.Module):
    """Logits processor that boosts the probability of constraint tokens."""

    def __init__(self, constraint_token_ids: Set[int], boost_factor: float = 2.0):
        super().__init__()
        # Store as a list rather than a tensor to avoid device issues
        self.constraint_token_ids = list(constraint_token_ids)
        self.boost_factor = boost_factor

    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor) -> torch.FloatTensor:
        """Boost scores for constraint tokens."""
        # Create tensor on the same device as scores
        constraint_token_ids = torch.tensor(
            self.constraint_token_ids,
            dtype=torch.long,
            device=scores.device
        )

        # Check if constraint token IDs are within the vocabulary size
        max_vocab_id = scores.size(1)
        valid_constraint_token_ids = constraint_token_ids[constraint_token_ids < max_vocab_id]

        # Apply boost to valid constraint tokens
        scores[:, valid_constraint_token_ids] *= self.boost_factor

        return scores


class ConstraintGuidedBART(nn.Module):
    """BART model with direct constraint integration."""

    def __init__(self, model_name="facebook/bart-large-cnn"):
        super().__init__()
        self.model = BartForConditionalGeneration.from_pretrained(model_name)
        self.tokenizer = BartTokenizer.from_pretrained(model_name)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model.to(self.device)

    def forward(self, input_ids, attention_mask, decoder_input_ids=None, labels=None):
        """Forward pass with standard BART."""
        return self.model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            decoder_input_ids=decoder_input_ids,
            labels=labels
        )

    def generate_with_constraints(
        self,
        input_ids,
        attention_mask,
        constraint_token_ids,
        num_beams=4,
        min_length=400,
        max_length=500,
        boost_factor=2.0
    ):
        """Generate text with constraint-guided beam search."""
        # Create logits processor with constraints
        logits_processor = LogitsProcessorList([
            MinLengthLogitsProcessor(min_length, self.tokenizer.eos_token_id),
            ConstrainedLogitsProcessor(constraint_token_ids, boost_factor)
        ])

        # Generate with constraints
        outputs = self.model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            num_beams=num_beams,
            min_length=min_length,
            max_length=max_length,
            logits_processor=logits_processor,
            early_stopping=True,
            no_repeat_ngram_size=3
        )

        return outputs

    def summarize(self, document, key_sentences, min_length_pct=0.75, max_length_pct=0.9):
        """Generate summary with constraints from key sentences."""
        # Calculate min and max length based on document length
        doc_length = len(document.split())
        min_length = int(doc_length * min_length_pct)
        max_length = int(doc_length * max_length_pct)

        print(f"Document length: {doc_length} words")
        print(f"Summary length range: {min_length}-{max_length} words")

        # Extract constraints
        constraint_extractor = ConstraintExtractor()
        constraints = constraint_extractor.extract_constraints(key_sentences)
        constraint_token_ids = constraint_extractor.get_constraint_tokens(constraints, self.tokenizer)

        # Tokenize input
        inputs = self.tokenizer(document, return_tensors="pt", max_length=1024, truncation=True)
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        # Generate summary
        print("Using constraint-guided beam search in ConstraintGuidedBART...")
        output_ids = self.generate_with_constraints(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            constraint_token_ids=constraint_token_ids,
            min_length=min_length,
            max_length=max_length
        )

        # Decode summary
        summary = self.tokenizer.decode(output_ids[0], skip_special_tokens=True)
        return summary


# Function to run the ConstraintGuidedBART
def run_constraint_guided_bart(document, extractive_key_sentences):
    """
    Run the ConstraintGuidedBART.

    Args:
        document: The full document text
        extractive_key_sentences: Key sentences extracted by Longformer

    Returns:
        Generated summary
    """
    print("\n--- Running ConstraintGuidedBART ---")
    summarizer = ConstraintGuidedBART()
    summary = summarizer.summarize(
        document=document,
        key_sentences=extractive_key_sentences,
        min_length_pct=0.75,
        max_length_pct=0.9
    )
    print("--- Finished ConstraintGuidedBART ---\n")

    return summary
```
